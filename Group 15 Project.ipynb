## CASE STUDY: MEDICAL INSURANCE COST PREDICTION

## Introduction

There has been an increase in the costs associated with healthcare services over the years. Bujang (2026) reported that healthcare spending in the United Sates is projected to reach $6.2 trillion by 2028, accounting for 20% of the GDP. In Europe, chronic diseases are responsible for 70%–80% of all healthcare costs while in Africa, non-communicable diseases are putting a lot of pressure on already limited healthcare resources. This rising cost therefore places a significant burden on government, individuals seeking treatment and insurance companies.

The correct prediction of medical expenses is important for effective resource allocation, policy-making and financial planning. Due to the complex relationship between patient characteristics, healthcare services and related costs, traditional method of estimation often miss the mark. For this reason, machine learning (ML) is now widely used to improve cost predictions, as well as identify the drivers of medical expenses (Kumar et al., 2024).

Even though cost prediction is important, an understanding of the factors that influence the predictions are also important. Key predictors such as demographic factors, lifestyle choices, and health conditions can provide valuable insights into the underlying causes of high medical expenses. Research has also revealed that variables such as age, body mass index (BMI), smoking status are among top predictors of healthcare costs; this shows their relationship with chronic diseases and other costly ailments (Kumar et al., 2024).


## Background to the study

Historically, health insurance premium determination relied heavily on manual underwriting, traditional actuarial tables, and broad demographic generalizations. While functional, these conventional methods often fail to capture complex, non-linear interactions between individual risk factors—such as how age interacts with smoking status or body mass index (BMI).

An insurance provider seeks to develop a reliable predictive model to estimate expected medical charges for its customers. The primary business challenge is to move beyond traditional, less granular pricing methods to a data-driven approach that accurately reflects individual risk profiles. The goal is to use personal and health-related information to forecast costs, thereby improving pricing accuracy and financial stability.

## Problem Statement

  The problem involves performing a detailed exploratory data analysis (EDA) to understand which demographic and lifestyle factors most significantly influence medical expenses. Subsequently, a robust regression model will be built and evaluated to predict individual healthcare insurance charges. The final model will be interpretable, allowing the insurance provider to understand the variables with the strongest influence on costs and to formulate actionable recommendations for risk assessment and pricing decisions.

## Objectives

- Conduct a thorough analysis of the dataset to understand the distribution of each feature and their relationships with the target variable. This includes visualizing correlations and identifying initial trends.
- Prepare the raw data for machine learning. This involves handling categorical variables through encoding techniques, scaling numerical features where necessary, and splitting the data into training and testing sets.
- Develop a predictive model using regression algorithms.
- Assess the predictive power of the model using standard regression metrics, such as R-squared (R²), Root Mean Squared Error (RMSE), and Mean Absolute Error (MAE).
- Interpret the final model to determine which features have the strongest influence on predicting charges.
- Based on the model's findings, formulate actionable recommendations for the insurance provider on risk assessment, pricing strategies, and the development of more personalized and competitive policies.

## Healthcare Insurance Terminologies

- Insurance policy: An insurance policy is a legally binding contract between an insurer (the insurance company) and a policyholder.
- Policy holder: A policyholder is the individual or business entity that legally owns an insurance policy.
- Premium: The monthly, quarterly, or annual amount paid by the policyholder to the insurance company to maintain coverage.
- Claim: A formal request submitted to the insurance company for payment of covered medical services. 
- Charges: The total amount billed for healthcare services by a provider.
- Deductible: The amount a policyholder must pay out-of-pocket for covered healthcare services before the insurance company begins to pay its share.
- Co-insurance: The policyholder's share of the costs of a covered healthcare service, calculated as a percentage or a fixed amount, which is paid after the deductible is met.
- Risk Assessment: The process of evaluating the likelihood of a policyholder filing a claim and the potential cost of that claim.
- Underwriting: The process used by insurance companies to evaluate the risk of insuring a person and determine the appropriate premium to charge.
- Beneficiary: The person who is covered by the health insurance policy and is eligible to receive benefits.
